#### Creating deck of cards as per requirement

The four sets in the list represent the different suites in the deck but as per requirement we are only considering it as a value.

```python
deck = [[2,3,4,5,6,7,8,9,10],
        [2,3,4,5,6,7,8,9,10],
        [2,3,4,5,6,7,8,9,10],
        [2,3,4,5,6,7,8,9,10]]
```

#### Total number of points in the deck
Reading each element of deck and adding the value to the total sum to get the total points in the deck.

```python
def count_value_in_list(l1):
    count_list = []
    for i in l1:
        for j in i:
            count_list.append(j)
    return (count_list,len(count_list))

print("__________________________________________________________________________")
print("Deck Check Block:")

total = 0
for curr_value in deck:
    for x in curr_value:
        total = total + x
print('Total sum of cards in the deck: ' + str(total))
print("__________________________________________________________________________")
```

#### Shuffling the cards in the deck randomly and distributing them equally to the players

Shuffling the cards multiple time before distributing them equally to the two players i.e 18 cards to each player.

```python
def distribute_deck(data):
    # step 1 shuffle
    random.shuffle(data)

    # Converging into one list and getting the length of elements in list
    shuff_list, len_shuff_list = count_value_in_list(data)
    
    # Step 2 Shuffle
    random.shuffle(shuff_list)

    # assignment
    play_1 = []
    play_2 = []
    count = 0
    for elem in shuff_list:
        if count < len(shuff_list)/2:
            play_1.append(elem)
        else:
            play_2.append(elem)
        count = count+1

    # last shuffle step
    random.shuffle(play_1)
    random.shuffle(play_2)
    
    return(play_1,play_2)

match = distribute_deck(deck)
print("Distribution in this match is: ",match)
print("\n")
tot1 = match[0]
print('Player 1 has ' + str(len(tot1)) + ' number of cards')
print('And they are: ' + str(tot1))
print('Sum of cards of player 1 is ', sum(tot1))
print("\n")
tot2 = match[1]
print('Player 2 has ' + str(len(tot2)) + ' number of cards')
print('And they are: ' + str(tot2))
print('Sum of cards of player 2 is ', sum(tot2))
print("\n")

```

```python
# in This function I'm taking the tuple with the two player lists and dividing them into two lists and selecting
# the first two cards of each player and removing the two cards selected from the list of both the players
def game_play(set1):
    first_set = set1[0]
    second_set = set1[1]
    play1_cards = first_set[0:2]
    play2_cards = second_set[0:2]
    sum1 = sum(play1_cards)
    sum2 = sum(play2_cards)
    print('sum1 is ' + str(sum1))
    print('sum2 is ' + str(sum2))
    print("---------------------------------------------")
    del first_set[0:2]
    del second_set[0:2]
    tot1 = first_set
    tot2 = second_set
    return sum1, sum2, tot1, tot2, play1_cards, play2_cards

## this function is used to pick the first two cards from the two players when a tie between two players occurs
def play_game(first_set, second_set):
    play1_cards = first_set[0:2]
    play2_cards = second_set[0:2]
    sum1 = sum(play1_cards)
    sum2 = sum(play2_cards)
    print('sum1 is ' + str(sum1))
    print('sum2 is ' + str(sum2))
    print("---------------------------------------------")
    del first_set[0:2]
    del second_set[0:2]
    tot1 = first_set
    tot2 = second_set
    return sum1, sum2, tot1, tot2, play1_cards, play2_cards

## This function is used whenver tie happens it picks the first two cards from each player from the play_game function
## and compares the sum and reacts accordingly

def tie_breaker(first_set, second_set, play1_cards, play2_cards):
    
    who_won = []
    
    sum1, sum2, tot1, tot2, p1_cards, p2_cards = play_game(first_set, second_set)
    
    for i in p1_cards[0:2]:
        play1_cards.append(i)
    for i in p2_cards[0:2]:
        play2_cards.append(i)
        
    if sum1 > sum2:
        who_won.append('player1')
    elif sum1 < sum2:
        who_won.append('player2')
    else:
        tot1, tot2, play1_cards, play2_cards, who_won = tie_breaker(tot1, tot2, play1_cards, play2_cards)
        
    return tot1, tot2, play1_cards, play2_cards, who_won 

## In this fuction we get the sum of first cards of both the players and compares them if one of them wins the match
## all the four cards are appended to the winning player.
## if there is a tie condition the loop then moves to tie-breaker and then the player who wins will get all the cards from the 
## tie-breaker.

def match_function(var):
    check_1 = 9 
    check_2 = 9
    count = 0

    while check_1 >= 0 and check_2 >= 0 and count < 25:
        sum_new1, sum_new2, tot_new1, tot_new2, play1, play2 = game_play(var)
        check_1 = len(tot_new1)
        check_2 = len(tot_new2)
        tie_list = []
        
        if sum_new1 > sum_new2:
            for i in play2:
                tot_new1.append(i)
            for i in play1:
                tot_new1.append(i)
            count = count + 1
            print('Player one has won the round: ' + str(count))
            print("---------------------------------------------")
            print('tot1 =' + str(tot_new1))
            print('tot2 =' + str(tot_new2))      
        elif sum_new2 > sum_new1:
            for i in play1:
                tot_new2.append(i)
            for i in play2:
                tot_new2.append(i)
            count = count + 1
            print('Player two has won the round: ' + str(count))
            print("---------------------------------------------")
            print('tot1 =' + str(tot_new1))
            print('tot2 =' + str(tot_new2))
        elif sum_new1 == sum_new2:
            print('TIE_BREAKER')
            print("==================")

            tot_new3, tot_new4, play_new3, play_new4, who_won1 = tie_breaker(tot_new1,tot_new2, play1, play2)
            print('For tie-breaker ' + str(who_won1) + ' won')
            for i in play_new3:
                tie_list.append(i)
            for i in play_new4:
                tie_list.append(i)
            print(tie_list)
            
            if "player1" in who_won1:
                count = count + 1
                print('Player one has won the round: ', str(count))            
                print("---------------------------------------------")
                for i in tie_list:
                    tot_new3.append(i)
            elif 'player2' in who_won1:
                count = count + 1
                print('Player two has won the round: ', str(count))
                print("---------------------------------------------")
                for i in tie_list:
                    tot_new4.append(i)
                
            print('tot1 = ',str(tot_new3))
            print('tot2 = ',str(tot_new4))
            
                    
        print("\n")
        ddv_1 = False
        ddv_2 = False
        ddv_3 = False
        score_1 = 0
        score_2 = 0
        score_3 = 0
        loss_score_play1 = 0
        loss_score_play2 = 0
        if tot_new1 == []:
            score_2 = score_2 + sum(tot_new2)
            print("Player 2 earns the entire set of cards from Player1 ")
            print("player 2 score is ", score_2)
            ddv_2 = True
            break
        elif tot_new2 == []:
            score_1 = score_1 + sum(tot_new1)
            print("Player 1 earns the entire set of cards from Player 2 ")
            print("Player 1 score is ", score_1)
            ddv_1 = True
            break
        elif count == 25:
            if sum(tot_new1) > sum(tot_new2):
                score_1 = score_1 + sum(tot_new1)
                loss_score_play2 = loss_score_play2 + sum(tot_new2)
                print("Player 1 score is ", score_1)
                print("Player 2 score is ", loss_score_play2)
                ddv_1 = True
                break

            elif sum(tot_new1) < sum(tot_new2):
                score_2 = score_2 + sum(tot_new2)
                loss_score_play1 = loss_score_play1 + sum(tot_new1)
                print("Player 2 score is ", score_2)
                print("Player 1 score is ", loss_score_play1)
                ddv_2 = True
                break
                
            elif sum(tot_new1) == sum(tot_new2):
                print('the final sum of the cards after 25th game are same as well')
                score_3 = score_3 + sum(tot_new2)
                ddv_3 = True
                break



    return ddv_1, ddv_2, ddv_3, score_1, score_2, score_3, loss_score_play1, loss_score_play2

```
